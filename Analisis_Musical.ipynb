```
from music21 import *
from IPython.display import Image, Audio
from midi2audio import FluidSynth

littleMelody = converter.parse("tinynotation: 3/4 c4 d8 f g16 a g f# c e a d e")
Image(filename=str(littleMelody.write('lily.png')))
```

```
def audio_out(ob):
    fs = FluidSynth('/usr/share/sounds/sf2/FluidR3_GM.sf2')
    fs.midi_to_audio(str(ob.write('midi')), 'output.wav')
    return Audio('output.wav')

def img_out(ob):
    return Image(filename=str(ob.write('lily.png')))

```

```
littleMelody.plot()
```

```
audio_out(littleMelody)
```

```
b = corpus.parse('bwv66.6')
img_out(b)
```
